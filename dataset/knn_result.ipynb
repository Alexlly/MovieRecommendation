```python
import tensorflow as tf
import numpy as np
from PIL import Image
import pandas as pd
import json

mostRatedMovies = pd.read_csv('./movie_rate_with_weight.csv')
mostRatedMovies
```

```python
len(mostRatedMovies['title'].unique())
```

```python
len(mostRatedMovies['userId'].unique())
```

```python
# Get a pivot view of each user's rating for each movie
moviePivot = mostRatedMovies.pivot_table(
    index=["title"],
    columns=["userId"],
    values=["rating"],
    fill_value=0,
)
```

```python
moviePivot
```

```python
import pandas as pd
import os
# create hdf5 file to save data
# Use h5 storage, which loads faster and can make the site serve better performance
h5_store = pd.HDFStore('moviePivot.h5', mode='w')
h5_store['moviePivot'] = moviePivot
# close
h5_store.close()
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd

# create csr_matrix matrix
movie_features_df_matrix = csr_matrix(moviePivot.values)
movie_features_df_matrix
```

```python
# fitting the final unsupervised model NearestNeighbors to 
# find the most similar movies of each ones using the whole dataset
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=11, n_jobs=-1)
model_knn.fit(movie_features_df_matrix)
```

```python
# generating recommendations related to a movie chosed
favoriteMovie = 'Iron Man (2008)'
query_index = moviePivot.index.get_loc(favoriteMovie)
distances, indices = model_knn.kneighbors(moviePivot.loc[favoriteMovie,:].values.reshape(1, -1), n_neighbors=11)
 
# printing the 10 most similar movies according to the kNN model
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(moviePivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, moviePivot.index[indices.flatten()[i]], distances.flatten()[i]))


```

```python
from sklearn.externals import joblib
joblib.dump(model_knn, "model_knn.m")
```

## Evaluation

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
# evaluating the kNeighborsRegressor model
X = mostRatedMovies[['userId', 'movieId']].values
y = mostRatedMovies['rating'].values
knn = KNeighborsRegressor(n_neighbors=11, n_jobs=-1)
knn.fit(X, y)
y_pred = knn.predict(X)
# for recommendation system, I use RMSE to measure the deviation between observed and true values
rmse = sqrt(mean_squared_error(y, y_pred))
print("Root-mean-square error = " + str(rmse))
```
